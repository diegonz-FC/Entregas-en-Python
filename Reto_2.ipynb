# 1.- La secuencia del cortador de pizza perezoso ( The Lazy Caterer's sequence).
Se trata un problema clásico de combinatoria y geometría que responde a la siguiente pregunta, ¿cuál es el número máximo de piezas en que se puede dividir una pizza utilizando n cortes rectos? Si tienes una pizza y haces cortes rectos, no necesariamente pasando por el centro y con la posibilidad de cruzarse en cualquier punto, ¿cuál es la mayor cantidad de rebanadas que puedes obtener?. Sabemos que

f(0)=1

f(1)=2

f(2)=4

f(3)=7

Con base en esta información, se puede utilizar la siguiente fórmula de recursión para n cortes.

f(n)=f(n−1)+n

Genere un programa en Python que muestre la secuencia de n cortes. Proponga un valor apropiado de cortes y justifique porqué escogió dicho valor.

```python
sec = [1] # Caso base de 0 cortes
def pizza():
  x = int(input('Ingrese un numero entero mayor igual que 0 de cortes: '))

  for x in range(1, x+1):
    sec.append(sec[-1]+x)
  print(sec)
pizza()

# Escogemos 5 cortes dado que escoger de 0 - 3 cortes es trivial y no tiene sentido (son muy pocas rebanadas), pero también escoger un número muy grande es bastante inadecuado ya que no tiene sentido partir una pizza del tal manera que se formen demasiados pedazos muy pequeños, además no se podrían visualizar bien las piezas.
```

# 2.- Serie de Madhava.
Utilice un ciclo for para estimar el valor de pi ( π ) a partir de los primeros 20 términos de la Serie de Madhava.

```python
i = 0
pi = 0
for i in range (0, 20):
  pi += (12**(1/2))*((-1)**(i) / ((2*i+1)*(3**(i))))
print(pi)
```
