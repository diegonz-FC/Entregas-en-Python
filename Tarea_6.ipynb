## 1. Suma de Matrices

```python
def suma(A, B):

  filasA = len(A)
  colsA = len(A[0])
  filasB = len(B)
  colsB = len(B[0])

  # Verificar que es posible la suma
  if filasA != filasB or colsA != colsB:
    print("Las matrices deben tener las mismas dimensiones para ser sumadas.")


  msuma = []

  # Sumar elemento a elemento
  for i in range(filasA):
    filanueva = []
    for j in range(colsA):
      sumar = A[i][j] + B[i][j]
      filanueva.append(sumar) # Va guardando el resultado de la suma por fila

      # Guardar cada fila en la matriz de la suma
    msuma.append(filanueva)

  return msuma


A = [[1, 2, 3], [4, 5, 6], [8, 10, 2]]
B = [[11, 15, 7], [8, 4, 9], [10, 4, 1]]


suma(A, B)
```

## 3. Traza de una matriz


```python
def traza(A):
  suma = 0

  filasA = len(A)
  colsA = len(A[0])

  # Verificar que es posible la multiplicación
  if filasA != colsA:
    raise Exception("Las matriz debe ser cuadrada para que la traza esté definida.")

    # Calcular la traza
  for i in range(filasA):
    for j in range(colsA):
      if i == j:
        suma += A[i][j]

  return suma


A = [[1, 2, 3], [4, 5, 6], [8, 10, 2]]

traza(A)
```

## 4. Transpuesta de una matriz

```python
def transpuesta(A):
  filasA = len(A)
  colsA = len(A[0])

  transp = []

    # El bucle exterior itera sobre las *columnas* de A.
    # Estas se convertirán en las *filas* de la matriz transpuesta (transp).

  for i in range(colsA):
    filanueva = []  # Inicia una lista vacía para cada columna

    for j in range(filasA):
        nuevoelemento = A[j][i] # Invierte columna con fila para ponerlo en las nuevas filas
        filanueva.append(nuevoelemento) # Agraga cada elemento a la fila nueva de la transpuesta.

        # Guarda cada fila nueva en la matriz transpuesta
    transp.append(filanueva)

  return transp

A = [[1, 2, 3], [4, 5, 6], [8, 10, 2]]

transpuesta(A)
```

## 5. Máximo y mínimo elemento de una matriz

```python
def maxmin(A):
  filasA = len(A)
  colsA = len(A[0])

  max = A[0][0] # Tomamos el primer elemento para empezar a comparar, puede ser cualquiera
  min = A[0][0]

  for i in range(filasA):
      for j in range(colsA): # Revisa todos los elementos de las columnas por cada fila
        val = A[i][j]  # Guarda el elemento a revisar para compararlo

        if val > max:
          max = val

        if val < min:
          min = val
  resultado = f'El máximo es {max} y el mínimo es {min}'
  return resultado

A = [[1, 2, 3], [4, 5, 6], [8, 10, 2]]

maxmin(A)
```

## 6. Suma de los cuadrados de los elementos de una matriz

```python
def sumacuadrado(A):

  filasA = len(A)
  colsA = len(A[0])

  suma = 0

  for i in range(filasA):
    for j in range(colsA):
      A[i][j] = (A[i][j])**2 # Eleva al cuadrado cada elemento
      suma += A[i][j] # Va sumando cada elemento


  return suma
A = [[1, 2, 3], [4, 5, 6], [8, 10, 2]]



sumacuadrado(A)
```

## 8. Funciones con una matriz aleatoria
8.- Genere una función que reciba como entradas las dimensiones de una matriz T[n, m] (3 <= n, m <=5).

Aseg urese de validar que sea cuadrada; de lo contrario, debe devolver el mensaje de error 'La matriz no
es cuadrada' y vuelva a preguntar las entradas.

Posteriormente reemplace los valores con n ́umeros aleatorios
e imprima:

```python
import random
def funciones():
  while True:
    n = int(input('Ingrese el número de filas: '))
    m = int(input('Ingrese el número de columnas: '))

    if (n < 3 or n > 5 or m < 3 or m > 5):
      print('Ingrese un número de filas y  columnas entre 3 y 5')

    elif n != m:
      print('La matriz no es cuadrada')

    else:
      break

  matriz = []

  # Generamos la matriz con datos aleatorios:
  for i in range(n):
    filanueva = []
    for j in range(m):
      termino = random.randint(1, 100)
      filanueva.append(termino)

    matriz.append(filanueva)

  nmatriz = np.array(matriz)

  print('La matriz es:')
  print(nmatriz)

  # Primer fila inversa
  filainversa = nmatriz[0][::-1] # Toma la primera fila y la invierte
  print(f'La primera fila de manera inversa es: ',filainversa)

  # Primer columna inversa
  columnainversa = nmatriz[:, 0][::-1] # De todas las filas, toma el primer elemento, luego lo invierte
  print(f'La primera columna de manera inversa es: ',columnainversa)
  #return matriz[0]

  # Primer elemento y último elemento.
  print('El primer elemento es: ',nmatriz[0][0])
  print('El último elemento es: ',nmatriz[-1][-1])

  # Valor máximo y mínimo, promedio de la diagonal, traza
  print('El valor máximo es:', np.max(nmatriz))
  print('El valor máximo es:', np.min(nmatriz))
  print('La traza de la matriz es: ', np.trace(nmatriz))
  print('El promedio de la diagonal es: ', round(((np.trace(nmatriz)) / len(nmatriz)), 2))

  # Los valores de T desde T[1,1] hasta T[5,5]
  listacompleta = []
  for fila in nmatriz:
    for elemento in fila:
      listacompleta.append(elemento)

  listaentera = [int(x) for x in listacompleta] # Para que no salga np.int64
  print('Los valores en orden desde T[1,1] a T[5, 5] son: ', listaentera)

  # Los valores desde T[1,5] hasta T[5,1]
  listacompleta2 = []
  for fila2 in nmatriz:
    for elemento2 in fila2[::-1]: # Invierte el orden
      listacompleta2.append(elemento2)

  listaentera2 = [int(x) for x in listacompleta2] # Para que no salga np.int64
  print('Los valores en orden inverso son: ', listaentera2)

  # Suma de la primera fila y la última columna
  sumarow = 0
  sumacol = 0
  for x in nmatriz[0]:
    sumarow += x
  for y in nmatriz[:, -1]: # Recorre todas las filas y accede al último elemento
    sumacol += y
  sumatotal = sumarow + sumacol
  print(f'La suma de la primera fila es {sumarow}, la de la última columna es {sumacol}, entonces la suma es {sumatotal}' )

  # Suma de elementos pares recorriendo filas
  sumapares = 0
  for fila in nmatriz:
    for elemento in fila:
      if elemento % 2 == 0:
        sumapares += elemento
  print('La suma de los pares es: ', sumapares)

  # Suma de elementos pares recorriendo columnas
  sumaimpares = 0
  for j in range(0, m):
    for i in range(0, n):
      elemento = nmatriz[i][j]
      if elemento % 2 != 0:
        sumaimpares += elemento
  print('La suma de los impares es: ', sumaimpares)

  # Promedio de filas
  for fila in nmatriz:
    total = 0   # Una variable para guardar la suma de cada fila
    for elemento in fila:
      total += elemento
      promediof = total / m   # Promedio de cada fila
    print('Los promedios de las filas son:', round(promediof, 2))

  # Promedio de columnas
  for j in range(0, m):
    total2 = 0    # Una variable por cada columna
    for i in range(0, n):
      elemento = nmatriz[i][j]
      total2 += elemento
      promediocol = total2 / n  # Promedio por columna
    print('Los promedios de las columnas son: ', round(promediocol, 2))

  # Antidiagonal y promedio
  # Invertimos la matriz
  matrizinvertida = np.fliplr(nmatriz)

  antidiagonal = np.diag(matrizinvertida)
  promedio = np.mean(antidiagonal)
  print('La antidiagonal es: ', antidiagonal)
  print('El promedio de la antidiagonal es: ', round(promedio, 2))

funciones()

```

## Ejercicio 1

```python
def edo(x, C):
  return (x-1) + C*np.exp(-x)

x = np.linspace(-10,4,200)

C1 = [1, 10, 20]
C2 = [-1, -10, -20]

# Con C mayor que 0
for a in C1:
  y = edo(x, a)
  plt.plot(x, y, label = 'C > 0')


# Con C mmenor que cero:
for a in C2:
  y = edo(x, a)
  plt.plot(x, y, label = 'C < 0')

# Con C igual a 0
y1 = edo(x, 0)
plt.plot(x, y1, label = 'C = 0')

font1 = {'family':'serif','color':'darkred','size':17}


plt.title('Problema Inicial', fontdict = font1)
plt.legend(loc = 'upper right', title = 'Valores de C')
plt.grid()
plt.show(True)
```
