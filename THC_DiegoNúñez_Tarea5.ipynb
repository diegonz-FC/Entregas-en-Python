## Ejercicio 4 - Números mayores a 100.
Escriba una función que lea un número entero (N) mayor que 100 y valide:

1. Si N es menor de 100 retorne: "Error en el número" y vuelva a preguntar la entrada.

2. Si N termina en 4, 7 ó 9; diga si es divisible por 3.

3. Si N termina en 1 ó 5, divida por 10 y retorne el valor entero,

4. En otro caso divida el número entre 3 y retorne su residuo.

```python
num = int(input('Ingrese un número entero mayor que 100: '))

def nm100(num):
  while num < 100:
    print('Error en el número.')
    num = int(input('Ingrese un número entero mayor que 100: '))

  if (num % 10 == 4 or num % 10 == 7 or num % 10 == 9): # Ver si termina en 4, 7, 9
    if num % 3 == 0:
      print(f'{num} es divisible por 3.')
    else:
      print(f'{num} no es divisible por 3.')

  elif (num % 10 == 1 or num % 10 == 5): # Ver si termina en 1, 5
    print(num // 10)

  else:
    print(num % 3)

nm100(num)
```

## Ejercicio 7 - Cálculo de la distancia entre dos puntos mediante la fórmula de Haversine.
Construya una función que permita calcular la distancia entre dos pares de coordenadas. Utiliza la librería math para las funciones trigonométricas y de raíz cuadrada.

Ejemplo de prueba. Coordenadas de Paris y Roma por latitud y longitud: París (48.9, 2.4) y Roma (41.9, 12.5). Resultado: distancia=1107 km.

Proponga 3 pares de coordenadas. Al menos uno debe ser respecto a la Ciudad de México (Latitud=19.43, Longitud=-99.13).

```python
def radianes(x):
  return x * (math.pi / 180)

def dHaversine(lat1, lon1, lat2, lon2):

  #Convertir a radianes las coordenadas
  lat1 = radianes(lat1)
  lat2 = radianes(lat2)
  lon1 = radianes(lon1)
  lon2 = radianes(lon2)

  # haversin
  haversin_lat = math.sin((lat2-lat1)/2)**2
  haversin_lon = math.sin((lon2-lon1)/2)**2

  R = 6378.1

  d = 2 * R * math.asin(math.sqrt(haversin_lat + math.cos(lat1) * math.cos(lat2) * haversin_lon))

  return print(f'La distancia entre los puntos es de {round(d,3)} km.')

# Paris - Roma
dHaversine(48.9, 2.4, 41.9, 12.5)

# CDMX - Whistler
dHaversine(19.43, -99.13, 50.116322, -122.957359)

# CDMX - Yokohama
dHaversine(19.43, -99.13, 35.443707, 139.638031)

# Valencia - Viena
dHaversine(39.47391, -0.37966, 48.20849, 16.37208)
```

## Ejercicio 8 - La serie de Fibonacci
Realice un programa que calcule la serie de Fibonacci y la imprima en pantalla, tenga en cuenta las restricciones de la variable de entrada.


```python
# La restricción es que la entrada sea mayor que 0.

def fibonacci(n):
  while n < 1:
    n = int(input('Ingrese una cantidad de elementos mayor que 0: '))

  if n == 1:
    serie = [0] # Def.

  if n == 2:
    serie = [0, 1, 1] # Def.

  if n > 2:
    serie = [0, 1]
    for i in range(2, n): # Def. recursiva
      serie.append(serie[i-1] + serie[i-2])

  print(serie)

n = int(input('Ingrese la cantidad de elementos de la serie de Fibonacci: '))
fibonacci(n)
```

## Ejercicio 10 - Operaciones vectoriales.
Utilice las listas de inicio para probar su código. Posteriormente, genere una función que reciba como entrada las listas (vectores) y haga las operaciones vectoriales.

Indique si algunos de los vectores son paralelos o perpendiculares.

```python
def vectors(a,b):
  # Separar y transformar en cadena
  a_str = a.split(",")
  b_str = b.split(",")

  # Transformar a floats
  a_flt = [float(x) for x in a_str]
  b_flt = [float(x) for x in b_str]
  print('El vector a es: ', a_flt)
  print('El vector b es: ', b_flt)

  # Producto escalar
  prodsc = sum(x*y for x, y in zip(a_flt, b_flt)) # Zip para que agarre uno con uno los datos en orden.
  print('El producto escalar es', prodsc)

  # Producto vectorial
  x = a_flt[1]*b_flt[2] - a_flt[2]*b_flt[1]
  y = a_flt[2]*b_flt[0] - a_flt[0]*b_flt[2]
  z = a_flt[0]*b_flt[1] - a_flt[1]*b_flt[0]
  prodv = [x, y, z]
  print('El producto vectorial es', prodv)

  # Paralelo, perpendicular.
  # En R3 si el producto escalar es 0 son ortogonales.
  if prodsc == 0:
    print('Los vectores son ortogonales.')

  elif prodv == [0, 0, 0]:
    print('Los vectores son paralelos.')

  else:
    print('Los vectores no son ortogonales ni paralelos')

a = input('Ingrese el vector a separado por comas: ')
b = input('Ingrese el vector b separado por comas: ')
vectors(a,b)
```
