# 1.- Juego de Piedra, Papel o Tijeras.
Genere una función que le permita jugar piedra, papel o tijera con la computadora, teniendo en cuenta tres posibles resultados: victoria, empate o derrota.

```python
def juego():
  print('Este es un juego de piedra, papel o tijeras')
  print('Sean:')
  print('1 = Piedra:')
  print('2 = Papel:')
  print('3 = Tijeras')

  seleccion = int(input('Escoge tu tirada: '))

  while seleccion > 3 or seleccion < 1:
    seleccion = int(input('Escoge nuevamente entre 1, 2, 3: '))

  if seleccion == '1':
     sf = 'piedra'
  elif seleccion == '2':
     sf = 'papel'
  else:
     sf = 'tijera'

  seleccioncomp = random.randint(1,3)
  if seleccioncomp == 1:
    sfc = 'piedra'
  elif seleccioncomp == 2:
    sfc = 'papel'
  elif seleccioncomp == 3:
    sfc = 'tijera'

  if sf == sfc:
    print('Empate')

  elif sf == 'piedra' and sfc == 'papel':
        print('¡Has perdido!')
  elif sf == 'papel' and sfc == 'tijera':
        print('¡Has perdido')
  elif sf == 'tijera' and sfc == 'piedra':
        print('¡Has perdido!')

  else:
    print('¡Has ganado!')

juego()


```
