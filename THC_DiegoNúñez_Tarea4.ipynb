# Fórmula de Herón de Alejandría.
Programe una función que calcule el área de un triángulo utilizando la fórmula de Herón.

```python
# Lados
a = float(input("Ingrese la medida del lado a: "))
b = float(input("Ingrese la medida del lado b: "))
c = float(input("Ingrese la medida del lado c: "))

def heron(a, b, c):
  print('Esta función calcula el área de un triángulo mediante la Fórmula de Herón de Alejandría.')

  #Desigualdad del triángulo, cantidad positiva:
  while not (a + b > c and a + c > b and b + c > a and a > 0 and b > 0 and c > 0):
    print("No existe un triangulo con con dichas medidas.")

    # Nuevos valores de la medida de los lados (Variables locales)
    a = float(input("Ingrese la medida del lado a: "))
    b = float(input("Ingrese la medida del lado b: "))
    c = float(input("Ingrese la medida del lado c: "))

  else:
    # Semiperímetro (Variable local s)
    s = (a + b + c) / 2

    # Área (Variable local)
    area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

    return round(area, 3)
print("El área del triángulo es:", heron(a, b, c))
```
